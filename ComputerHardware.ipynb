In this data I want to use Z-score to detect and handle outliers and also we can use IQR

```python
from pandas.core.internals.construction import dataclasses_to_dicts
# Create a DataFrame
df = pd.DataFrame(dataset)

# Select only numeric columns for outlier detection
numeric_columns = df.select_dtypes(include=[np.number])

# Calculate Z-Scores for numeric columns
z_scores = np.abs(zscore(numeric_columns))

# Define threshold for outlier detection (e.g., Z-Score > 2)
threshold = 2

# Identify outlier rows
outlier_rows = (z_scores > threshold).any(axis=1)

# Create DataFrame without outlier rows
df_no_outliers = df[~outlier_rows]

print("Original DataFrame:")
print(df)
print("\nDataFrame without Outliers:")
print(df_no_outliers)
csv_filename = 'data_without_outliers.csv'
df_no_outliers.to_csv(csv_filename, index=False)
```

Boxplot also show some outliers. Now I want to try IQR method for this column to see any changing happen or not in original data

```python
# Create the original DataFrame
original_df = pd.DataFrame(dataset)

# Calculate IQR for numeric columns
Q1 = original_df['Machine Cycle Time in nanoseconds'].quantile(0.25)
Q3 = original_df['Machine Cycle Time in nanoseconds'].quantile(0.75)
IQR = Q3 - Q1

# Define upper and lower bounds for outlier detection
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify and remove outlier rows
df_no_outliers = original_df[
    (original_df['Machine Cycle Time in nanoseconds'] >= lower_bound) &
    (original_df['Machine Cycle Time in nanoseconds'] <= upper_bound)
]

# Save DataFrame without outliers to a CSV file
csv_filename = 'data_without_outliers_IQR.csv'
df_no_outliers.to_csv(csv_filename, index=False)

print(f"DataFrame without outliers saved to '{csv_filename}'")
```
